## PCA

Per últim, aplicarem PCA al nostre dataset per a veure si podem reduir la dimensionalitat el suficient com a per poder representar-lo a 2D, i per a poder obtenir més informació sobre la correlació dels nostres atributs.

```python
data = df.values

x = data[:, :-1]  # Caracteristiques (x)
y = data[:, -1]  # Atribut y --> variable dependent del nostre model

pca = PCA(n_components=2)
new_X = pca.fit_transform(x)
exp_var = pca.explained_variance_

plt.bar(np.arange(2),exp_var/np.sum(exp_var))
plt.title('Explained variance')
```

El nostre PCA de dos components ens ha donat una variancia molt alta i altre molt baixa. Anem a representar la contribució de cada atribut als components per a veure que es el que està passant:

```python
components = pca.components_

for i,vector in enumerate(components):
    print(vector)
    plt.bar(np.arange(30),vector)
    plt.title(f'Contributions for component {i}')
    plt.show()


plt.bar(np.arange(30),np.sum(np.abs(components),axis = 0))
plt.title(f'Total contributions')
plt.show()
```

El component 0 (El qual era el que té la variancia més alta), es format principalment per l'atribut absencies, el qual diu quants dias ha faltat a classe un alumne. Per una part, aquest atribut ja ens había donat una correlació molt alta i té sentit que sigui un dels atributs principals del nostre PCA, però el seu valor sembla molt alt.
L'atribut absencies és l'únic atribut continu, i no s'ha standaritzat, anem a standaritzar les dades per a veure si cambia la nostra reducció de dimensions:

```python
# copy the data
df_z_scaled = df.copy()
  
# apply normalization technique to Column 1
column = 'absences'
df_z_scaled[column] = (df_z_scaled[column] - df_z_scaled[column].mean()) / df_z_scaled[column].std()   

data = df_z_scaled.values

x = data[:, :-1]  # Caracteristiques (x)
y = data[:, -1]  # Atribut y --> variable dependent del nostre model

pca = PCA(n_components=2)
new_X = pca.fit_transform(x)
exp_var = pca.explained_variance_

plt.bar(np.arange(2),exp_var/np.sum(exp_var))
plt.title('Explained variance')
```

Normalitzant les dades, hem aconseguit una variancia bastant equitativa, els dos components s'apropen molt al 50%.
Anem a veure que ha cambiat atribut a atribut:

```python
components = pca.components_

for i,vector in enumerate(components):
    print(vector)
    plt.bar(np.arange(30),vector)
    plt.title(f'Contributions for component {i}')
    plt.show()


plt.bar(np.arange(30),np.sum(np.abs(components),axis = 0))
plt.title(f'Total contributions')
plt.show()
```

La contribució dels nostres atributs als components s'ha estabilitzat i ara obtenim uns 10 atributs que superen el 0.2 de variancia. No obtenim cap atribut que destaqui excessivament de la resta, i podem veure com el nostre PCA te en compte sobre tot la següent informació:
- Consum d'alcohol als caps de setmana (Walc)
- Educació de la mare (Medu)
- Educació del pare (Fedu)
- Treball de la mare (Mjob)

Abans de finalitzar la secció del PCA, anem a veure com queden representades les dades gràficament en 2D.

```python
plt.scatter(new_X[:,0],new_X[:,1],c = y)
plt.show()
```

```python
# creating a Dataframe object
df3 = df.copy()

# Creating PCA dataframe
df4 = pd.DataFrame({'PC0': new_X[:,0], 'PC1': new_X[:,1], 'G3': df[df.columns[-1]]})

# apply normalization technique to Column 'absences'
column = 'absences'
df3[column] = (df3[column] - df3[column].mean()) / df3[column].std()   

cols1= ['age','Medu','Fedu','failures','higher','G3']
df1 = df3[cols1]

cols2= ['age','Medu','Fedu','failures','higher','sex','address','famsize','Mjob','reason','traveltime','studytime','schoolsup','paid','internet','romantic','goout','G3']
df2 = df3[cols2]
```

# 7. Creació de model

Per al nostre regressor lineal utilitzarem la funció SGDRegressor de sklearn. Primer farem una versió bàsica on compararem els tres datasets obtinguts, i amb els resultats, agafarem el millor dataset per a millorar els parametres del nostre model i obtenir els resultats finals.

```python
## DATASET 1

data = df1.values

x = data[:, :-1]  # Caracteristiques (x)
y = data[:, -1]  # Atribut y --> variable dependent del nostre model

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.2, random_state=42)

sgdr = SGDRegressor()
sgdr.fit(xtrain, ytrain)

score = sgdr.score(xtrain, ytrain)
print("R-squared:", score)


print("CV mean score: ", np.mean(cross_val_score(sgdr, x, y, cv=5,scoring='r2')))

ypred = sgdr.predict(xtest)

mse = mean_squared_error(ytest, ypred)
print("MSE: ", mse)
print("RMSE: ", mse**(1/2.0))
```

```python
## DATASET 2

data = df2.values

x = data[:, :-1]  # Caracteristiques (x)
y = data[:, -1]  # Atribut y --> variable dependent del nostre model

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.2,random_state=42)

sgdr = SGDRegressor()
sgdr.fit(xtrain, ytrain)

score = sgdr.score(xtrain, ytrain)
print("R-squared:", score)

from sklearn.model_selection import KFold

for n in range(5):
    k_fold = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(sgdr, x, y, cv=k_fold)

#print("CV mean score: ", np.mean(cross_val_score(sgdr, x, y, cv=5,scoring='r2')))

print("CV mean score: ", np.mean(scores))

ypred = sgdr.predict(xtest)

mse = mean_squared_error(ytest, ypred)
print("MSE: ", mse)
print("RMSE: ", mse**(1/2.0))
```

### Regressor lineal simple

Per a la regressió lineal bàsica utilitzarem el model LinearRegression de la llibreria sklearn.

```python
# Definim el regressor lineal
def regression(x, y):
    # Creem un objecte de regressió de sklearn
    regr = LinearRegression()

    # Entrenem el model per a predir y a partir de x
    regr.fit(x, y)

    # Retornem el model entrenat
    return regr

def comprobar_regresio(regr, x, y):
    predicted = regr.predict(x)
    MSE = mean_squeared_error(y, predicted)
    r2 = r2_score(y, predicted)


    return predicted, MSE, r2

def separar_objectiu(data, objectiu_idx):
    """funcio per a separar el atribut objectiu de les nostres dades"""
    x = np.delete(data, objectiu_idx, axis=1)
    y = data[:, objectiu_idx]
    return x, y

def dividir_train_test(dataset, perc):
    idx = dataset.sample(frac=perc).index

    train_dataset = dataset[dataset.index.isin(idx)]
    test_dataset = dataset[~dataset.index.isin(idx)]
    return train_dataset, test_dataset

def generar_train_test(dataset, objective_parameter):
    # Separem les dades de test i de train
    train_dataset, test_dataset = dividir_train_test(dataset, 0.7)

    # #obtenim x, y (objectiu 7: density)
    train_x, train_y = separar_objectiu(train_dataset.values, objective_parameter)
    test_x, test_y = separar_objectiu(test_dataset.values, objective_parameter)
    
    return train_x, train_y, test_x, test_y

def mean_squeared_error(y1, y2):
    # comprovem que y1 i y2 tenen la mateixa mida
    assert(len(y1) == len(y2))
    mse = 0
    for i in range(len(y1)):
        mse += (y1[i] - y2[i])**2
    return mse / len(y1)


def test_linear_regression(dataset, objective_parameter, chosen_param, graph=True):
    # Separem les dades de test i de train
    train_x, train_y, test_x, test_y = generar_train_test(dataset, objective_parameter)

    regr = regression(train_x[:, chosen_param].reshape(train_x.shape[0], 1), train_y)

    predicted, MSE, r2 = comprobar_regresio(regr, test_x[:, chosen_param].reshape(test_x.shape[0], 1), test_y)
    print("Mean squeared error: ", MSE)
    print("R2 score: ", r2)

    if graph:
        plt.figure()
        ax = plt.scatter(test_x[:, chosen_param], test_y)
        plt.plot(test_x[:, chosen_param].reshape(test_x.shape[0], 1), predicted, 'r')

    return regr, MSE, r2
```

En el cas dels dataset amb alta correlació, no podem utilitzar tots els atributs a l'hora, per tant mirarem quin és el que dona millors resultats:

```python
MSEs = []
r2s = []
objectiu_idx = 5
cols = np.delete(df1.columns, objectiu_idx)

for i, col in enumerate(cols):
    print("\t", col, ":")
    regr, mse, r2 = test_linear_regression(df1, objectiu_idx, i, False)
    
    MSEs.append(mse)
    r2s.append(r2)

MSEs = np.array(MSEs)
r2s = np.array(r2s)
plt.figure(figsize=(20,7)) 
plt.scatter(cols, MSEs, label='mse')
#plt.scatter(cols, r2s, label='r2')
plt.legend()
plt.show()
```

Dels 5 atributs amb millor correlació, el que pitjors resultats ha donat ha sigut 'higher' i 'age', els quals indiquen si volen fer estudis superiors, i la seva edat. Podem veure com els atributs més decisius per a la nota de l'estudiant son les seves asignatures suspeses, i l'educació dels seus pares.

### Random Forest

Per a implementar el Random Forest, hem de passar el nostre objectiu a predir a un tipus categoric. Podem utilitzar el sistema de notes dels Estats Units com a referencia:


```python
grades = {
    18: "A",
    16: "B",
    14: "C",
    12: "D",
    0: "F",
}

def grade_mapping(value):
    for key, letter in grades.items():
        if value >= key:
            return letter

letter_grades = df1["G3"].map(grade_mapping)
df1["G3"] = pd.Categorical(
    letter_grades, categories=grades.values(), ordered=True
)

df1
```

```python
# Funcions d'utilitat

def mostrar_validacio(modelo, X_train, y_train, X_test, y_test, draw_curves=False, classifier="Unknown"):
    #Predicciones train
    print("Predicting train values...")
    predicciones = modelo.predict(X_train)
    
    print("Predicted train values")

    accuracy = accuracy_score(
                y_true    = y_train,
                y_pred    = predicciones,
                normalize = True
            )
    print("")
    print(f"El accuracy de train es: {100*accuracy}%")
    f1 = f1_score(
                y_true    = y_train,
                y_pred    = predicciones,
                average   = 'micro'
            )
    print(f"El f1 score de train es: {100*f1}%")
    print("")

    #Predicciones test
    print("Predicting test values...")
    predicciones = modelo.predict(X_test)
    print("Predicted test values")

    accuracy = accuracy_score(
                y_true    = y_test,
                y_pred    = predicciones,
                normalize = True
            )
    print("")
    print(f"El accuracy de test es: {100*accuracy}%")
    f1 = f1_score(
                y_true    = y_test,
                y_pred    = predicciones,
                average   = 'micro'
            )
    print(f"El f1 score de test es: {100*f1}%")
    print("")
    
    return y_test, predicciones
    
    
```

```python
# Implementació Random Forest

train_x, train_y, test_x, test_y = generar_train_test(df1, 5)

modelo = RandomForestClassifier(n_estimators=100,random_state=42)
print("Created model")
print("Training model...")
modelo.fit(train_x, train_y)
print("Trained model :D")

y_test, predicciones = mostrar_validacio(modelo, train_x, train_y, test_x, test_y, False, "Random Forest")

cm = confusion_matrix(y_test, predicciones)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['A','B','C','D','F'])
disp.plot()
```

L'execució de l'algorisme Random Forest ens ha donat un 67,15% d'accuracy al training set, i un 54% d'accuracy al test.
Observant la matriu de confusió podem veure com el model pot predir bé els alumnes amb mala nota, pero es incapaç de predir les notes superiors a B.
La possible teoria a això és que la dificultat en pujar nota entre F i D, i la dificultat de pujar nota entre B i A és molt més alta, i el nostre model agafa una tendencia a les notes baixes. També hem de tenir en compte que la nota F té un rang molt més alt de notes que la resta.
